```python
#importing the libraries 
# For DSA we need to load in some tools for it to use, as well as an LLM for the brain/reasoning
from langchain.agents import load_tools  # This will allow us to load tools we need
from langchain.agents import initialize_agent
from langchain.agents import (
    AgentType,
)  # We will be using the type: ZERO_SHOT_REACT_DESCRIPTION which is standard
from langchain.llms import OpenAI
```

```python
# setup the openai and serp api keys
import os 
from secret_keys import openapi_key, serpapi_key
os.environ['OPENAI_API_KEY'] = openapi_key
os.environ['SERPAPI_API_KEY']= serpapi_key
```

```python
# For OpenAI we'll use the default model for DaScie
llm = OpenAI()
tools = load_tools(["wikipedia", "serpapi", "python_repl", "terminal"], llm=llm)
# We now create DSA using the "initialize_agent" command.
dsa = initialize_agent(
    tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True
)
```

```python
#Testing out Agent skills
#  Let's see how well DaScie can work with data on Wikipedia and create some data science results.
```

```python
# Let's see how well DaScie can work with data on Wikipedia and create some data science results.
dsa.run('Create a detailed dataset , following the name, team and score.I want you to get the data of top 10 players in ipl in 2023 based on the score in ipl, and make a barchart having the names of the players and the scores.')
```

```python
# Let's try to improve on these results with a more detailed prompt.
dsa.run(
    "Create a detailed dataset (DO NOT try to download one, you MUST create one based on what you find) on the GDP of India from 2014 - 2019 and do some analysis and give me some cool visuializations to understand the trend, you can give the line graph from the year 2014 - 2019, based on the Growth, use seaborn to plot the graphs."
)
```

```python
# Now we will use some local data for DSA to analyze.
# For this, we'll change dsa's configuration so it can focus on pandas analysis of some world data.

from langchain.agents import create_pandas_dataframe_agent
import pandas as pd

data_df = pd.read_csv("ds_salaries.csv")
# world_data
dsa = create_pandas_dataframe_agent(
    OpenAI(temperature=0), data_df, verbose=True
)
```

```python
# Let's see how well DSA does on a simple request.
dsa.run("Analyze this data, tell me any interesting trends. Make some pretty plots.")
```

```python
# Cell 7 - Not bad! Now for something even more complex.... can we get our LLM model to do some ML!?
dsa.run( 
    "Train a random forest regressor to predict salary using the most important features. and give me the plot with the model, Show me the what variables are most influential to this model"
)
```
